```python
train_df= pd.read_csv('train.csv')
test_df= pd.read_csv('test.csv')
# sample_df= pd.read_csv('/kaggle/input/spaceship-titanic/sample_submission.csv')
train_df.head()
```

```python
train_df= train_df.dropna(subset=["HomePlanet", "CryoSleep", "Cabin", "Destination", "Age", "VIP", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"])
```

```python
train_df=train_df.drop(['Cabin',],axis=1)
train_df.head()
```

```python
train_dummie= pd.get_dummies(train_df, columns=['HomePlanet', 'Destination'])
train_dummie.head()
```

```python
train_dummie.columns
```

```python
column_names = {'HomePlanet_Earth': 'Earth', 
                    'HomePlanet_Europa': 'Europa', 
                    'HomePlanet_Mars': 'Mars', 
                    'Destination_55 Cancri e': 'Cancri', 
                    'Destination_PSO J318.5-22': 'PSO',
                    'Destination_TRAPPIST-1e': 'Trappist'}

train_dummie.rename(columns=column_names,inplace=True)
```

```python
train_dummie.head()
```

```python
train_dummie['CryoSleep'] = train_dummie['CryoSleep'].astype(int)
train_dummie['VIP'] = train_dummie['VIP'].astype(int)
train_dummie['Transported'] = train_dummie['Transported'].astype(int)
train_dummie['Earth'] = train_dummie['Earth'].astype(int)
train_dummie['Europa'] = train_dummie['Europa'].astype(int)
train_dummie['Mars'] = train_dummie['Mars'].astype(int)
train_dummie['Cancri'] = train_dummie['Cancri'].astype(int)
train_dummie['PSO'] = train_dummie['PSO'].astype(int)
train_dummie['Trappist'] = train_dummie['Trappist'].astype(int)
train_dummie.head()


```

```python
x_train= train_dummie.drop('Transported',axis=1)
```

```python
x_train.head()
```

```python
y_train=train_dummie['Transported']
```

```python
test_dummie['CryoSleep'] = test_dummie['CryoSleep'].astype(int)
test_dummie['VIP'] = test_dummie['VIP'].astype(int)
test_dummie['Earth'] = test_dummie['Earth'].astype(int)
test_dummie['Europa'] = test_dummie['Europa'].astype(int)
test_dummie['Mars'] = test_dummie['Mars'].astype(int)
test_dummie['Cancri'] = test_dummie['Cancri'].astype(int)
test_dummie['PSO'] = test_dummie['PSO'].astype(int)
test_dummie['Trappist'] = test_dummie['Trappist'].astype(int)
test_dummie.head()


```
